```python
### importing libraries 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as no
import pandas as pd
import geopandas as gpd
from geopandas.tools import geocode
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
```

```python
## importing dataset
house_data = pd.read_csv('kc_house_data.csv')
house_data.head()
```

```python
## data information
house_data.info()
```

There are no missing values in our dataset, which is compound of 21613 rows and 21 columns. For now, there is no need to  formatting the data. 

The data source doesn't have a description for the variables in each column, but this is intiutive by looking at the column names, except with the column 'view'. I'll try to uncover what is this column telling about the houses.

```python
## some statistics for continuous variables
data_filter = house_data[['price', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_living15','sqft_lot15']]
data_filter.describe()
```

* The maximum value in each column is far from the 75% percentile, what means that we may have outliers in this columns or that the data is just spread. 
* These columns may be right skewed since the mean is greater that the median in each of them.
* According to the *sqft_basement* column at least half of the houses don't have a basement. 

```python
## continuius data correlation matrix
data_corr = data_filter.corr()

## heatmap
plt.figure(figsize=(15,6))
sns.heatmap(data=data_corr, annot=True)
plt.title('Continuous Data Correlation', fontsize=15, weight='bold')
```

* Columns *sqft_living*, *sqft_above* and *sqft_living15* are strongly correlated one another and with the target column (*price*). Before training and fitting the regression models I'll remove those columns from the data for having the models to "think" more. 
* It's also advisable to remove column *sqft_lot15* since it is highly correlated with the column *sqft_lot*

```python
## Price distribution
plt.figure(figsize=(15,6))
sns.distplot(a=house_data.price)
plt.title('Price Distribution', fontsize=15, weight='bold')
plt.xlabel('Price', fontsize=12)
```

As I stated earlier the price data is right skewed with some houses having really high prices compared to the most. 

```python
## Bedrooms count

sns.catplot(x='bedrooms', kind='count', data=house_data, palette='magma')
```

* The 'mode' of the dataset are houses with 3 bedrooms followed by houses with 4 bedrooms. 
* Within the data there is even a house with 33 bedrooms!

```python
## average house price by number of bedrooms
house_data = house_data.astype({'bedrooms':'category'})
table1 = house_data.groupby('bedrooms')['price'].mean().reset_index()
table1 = table1.sort_values(by='price', ascending=False)

### barplot
plt.figure(figsize=(15,6))
plt.title('Average house price by # Bedrooms', fontsize=15, weight='bold')
order = table1['bedrooms']
ax = sns.barplot(x='price', y='bedrooms', data=table1, palette='Greens', order=order)

for p in ax.patches:
    ax.annotate('USD {:.0f}'.format(p.get_width()), (p.get_x()+0.9*p.get_width(), p.get_y()+p.get_height()-0.2))
```

Surprisingly, the house with 33 bedrooms is not most expensive. This plot shows us that the number of bedrooms is not a main factor in order to figure out the house prices. 

Now, we can see that the type of the column *bathrooms* is 'float64' when it should be integer (we don't say that we have 2 and a half bathrooms in our house, do we?). So, I'll use the round function to round the number to the nearest integer and then I'll convert the column to category as I did with the bedrooms column. The same procedure will be applied to the *floors* column. 

```python
## rounding
house_data = house_data.round({'bathrooms':0, 'floors':0})

##setting as category
house_data = house_data.astype({'bathrooms':'category', 'floors':'category'})

house_data[['bathrooms', 'floors']].head(10)
```

```python
## houses bathrooms
print(house_data.bathrooms.unique(),'\n')

## house floors
print(house_data.floors.unique())
```

Where we have houses with up to 8 bathrooms and 4 floors. Let's see the average price for each of then.

```python
## house price by bathrooms
table1 = house_data.groupby('bathrooms')['price'].mean().reset_index()
table1 = table1.sort_values(by='price', ascending=False)

## barplot
plt.figure(figsize=(15,6))
plt.title('House Average Price by Bathrooms', fontsize=15, weight='bold')
ax = sns.barplot(x='bathrooms', y='price', data=table1, palette='Blues', order=table1['bathrooms'])

for p in ax.patches:
    ax.annotate('USD {:.0f}'.format(p.get_height()), (p.get_x(), p.get_height()), weight='bold')
```

House prices clearly depend on the number of bathrooms within them.

```python
### price behavior by floor number
figs, axes=plt.subplots(2,2,figsize=(15,8))
plt.tight_layout(pad=6)

## one floor
ax1 = sns.kdeplot(data=house_data[(house_data.floors == 1)]['price'], color='yellow', ax=axes[0,0])
ax1.set_title('One floor')

## Two floor
ax2 = sns.kdeplot(data=house_data[(house_data.floors == 2)]['price'], color='orange', shade=True, ax=axes[0,1])
ax2.set_title('Two floors')

## Three floor
ax3 = sns.kdeplot(data=house_data[(house_data.floors == 3)]['price'], color='purple', shade= True, ax=axes[1,0])
ax3.set_title('Three floors')

## Four floor
ax4 = sns.kdeplot(data=house_data[(house_data.floors == 4)]['price'], color='gray', ax=axes[1,1])
ax4.set_title('Four floors')
```

* There is not difference in the price distribution of houses according to the floor number. 
* Houses with one and two floors contain data with high prices compared to houses with three and four floors.

```python
### sqft_living vs price
fig, [axes1, axes2]=plt.subplots(1,2,figsize=(10,6))
plt.tight_layout(pad=5)

## sqft_living
ax1 = sns.scatterplot(x='sqft_living', y='price', data=house_data, ax=axes1)
ax1.set_title('Price vs sqft_living', weight='bold')

## sqft_living15
ax2 = sns.scatterplot(x='sqft_living15', y='price', data=house_data, ax=axes2)
ax2.set_title('Price vs sqft_living15', weight='bold')
```

* The behavior of both variables against the price column is basically the same. 
* Both of them are correlated as I stated earlier
* House prices increase as the living room squared feet raises

Let's plot the same graphs for the house squared feet columns. 

```python
### sqft_lot vs price
fig, [axes1, axes2]=plt.subplots(1,2,figsize=(10,6))
plt.tight_layout(pad=5)

## sqft_living
ax1 = sns.scatterplot(x='sqft_lot', y='price', data=house_data, ax=axes1, color='red')
ax1.set_title('Price vs sqft_lot', weight='bold')

## sqft_living15
ax2 = sns.scatterplot(x='sqft_lot15', y='price', data=house_data, ax=axes2, color='red')
ax2.set_title('Price vs sqft_lot5', weight='bold')
```

Both colum behave the same againt the price column, althought there is not a clear correlation between the house prices and their squared feet. 

```python
print(house_data.waterfront.unique())
print(house_data.view.unique())
print(house_data.condition.unique())
```

```python
## waterfront
ax1 = (house_data.waterfront.value_counts()*100/len(house_data)).plot.pie(autopct='%.2f%%', figsize=(10,6))
```

Just 162 houses have a waterfront. Where the average price for them is the following:

```python
## waterfront prices
table1 = house_data.groupby('waterfront')['price'].mean().reset_index()
table1
```

Houses with waterfront are more expensive than the others. 

```python
##barplot
plt.figure(figsize=(10,6))
sns.barplot(x='view', y='price', hue='waterfront', data=house_data, ci=None, palette='Set2')
plt.title('Average Price by View', fontsize=15, weight='bold')
```

* The more the view, the higher the house price. 
* Houses with a view of 2 and waterfront present the highest average prices. 
* Houses with a view of 0 don't have *waterfront* values. 
* From this we can say that the *view* column refers to the visibility from the house to the street. 

```python
### house condition
figs, axes=plt.subplots(1,2,figsize=(15,6))
plt.tight_layout(pad=5)

## type
house_data = house_data.astype({'condition':'category'})

##count
condition_count = house_data.groupby('condition')['price'].size().reset_index()

## condition count
ax1 = sns.barplot(x='condition', y='price', data=condition_count, palette='Oranges', ax=axes[0])
ax1.set_title('House Condition Count')
ax1.set_ylabel('Count')

## condition price
condition_price = house_data.groupby('condition')['price'].mean().reset_index()
condition_price = condition_price.sort_values(by='price', ascending=False)

ax2=sns.barplot(y='condition', x='price', data=condition_price, palette='Oranges', order=condition_price['condition'], ax=axes[1])
ax2.set_title('House Condition vs Price')
```

* Houses with a average condition (3) are the mode in the dataset. 
* Higher prices are paid for houses with a condition of 5, while the is not much different in the amount of money paid for houses with conditions 3 and 4, and conditions 1 and 2.
