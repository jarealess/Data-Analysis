```python
house_data.yr_renovated.unique()
```

I'll create the columns 'antiquity' and 'yr_from_renovation', from columns *yr_built* and *yr_renovated* in order to convert the year data to numeric values. 

Column *yr_renovated* has 0 values what may indicate those houses haven't been renovated. 

```python
## sale year
house_data['sale_year'] = house_data['date'].apply(lambda x: x[0:4])
house_data = house_data.astype({'sale_year':'int64'})

## adding antiquity column
house_data['antiquity'] = house_data['sale_year'] - house_data['yr_built']

##years from renovation
house_data['yr_from_renovation'] = abs(house_data['yr_built'] - house_data['yr_renovated'])
house_data['yr_from_renovation'] = house_data['yr_from_renovation'].apply(lambda x: 0 if x > 1000 else x)
house_data['yr_from_renovation'] = house_data['antiquity'] - house_data['yr_from_renovation']
```

```python
house_data[['antiquity', 'yr_from_renovation', 'yr_built', 'yr_renovated', 'sale_year']].head(10)
```

The column yr_from_renovation comprises the 'real age' of the house, since it is a mixed between antiquity and year of the last renovation. Let's see the behavior of this column against the price column.

```python
### yr_from_renovation vs price
plt.figure(figsize=(15,6))
sns.scatterplot(x='yr_from_renovation', y='price', data=house_data)
plt.title('House Real Antiquity vs Price', fontsize=15, weight='bold')
```

There is not clear relationship between the house prices and their real antiquity. 

### **Model Training and Fitting**

```python
## importing libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

Now, let's format the data set in an adequate form for fitting a ML model. In this step I'm going to drop those columns that don't contribute to the price prediction and also those columns that are highly correlated with the target column or with another column (like 'sqft_lot15' column).

```python
## house_data_mod
house_data_mod = house_data.drop(['id', 'date', 'sqft_living', 'sqft_above', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15','sqft_lot15', 'sale_year'], axis=1)
house_data_mod = house_data_mod.astype({'bedrooms': 'category', 'bathrooms': 'category', 'floors': 'category', 'waterfront': 'category', 
                                        'view': 'category', 'condition': 'category', 'grade': 'category'})
house_data_mod.head()
```

Now, let's split the data leaving 20% for testing.

```python
### separating column
target = house_data_mod['price']
feature = house_data_mod.drop(['price'], axis=1)

##splitting the data into 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=0)
```

Scaling and training the data... 

```python
## setting the scaler
scaler = StandardScaler()

## fitting training set
X_train_std = scaler.fit_transform(X_train)

## apply transform to test set
X_test_std = scaler.transform(X_test)
```

```python
## Linear Regression
lr = LinearRegression()
lr.fit(X_train_std, y_train)

## Random forest
rf = RandomForestRegressor()
rf.fit(X_train_std, y_train)

## Lasso
la = Lasso()
la.fit(X_train_std, y_train)
```

Precision of each model for predicting the house prices:

```python
## score Linear Regression
print('Score of {:.2f} was obtained with Linear Regression model'.format(lr.score(X_test_std, y_test)))

# score Random Forest Regressor
print('Score of {:.2f} was obtained with Random Forest Regressor model'.format(rf.score(X_test_std, y_test)))

## score Lasso
print('Score of {:.2f} was obtained with Lasso model'.format(la.score(X_test_std, y_test)))
```

A very low score was obtained with the three models. Let's use **sklearn PolinomialFeature** to see if it is possible to improve the models. 

```python
### separating column
Y = house_data_mod['price']
X = house_data_mod.drop(['price'], axis=1)

## polinomial features

poly = PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)

##splitting the data into 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X_poly, Y, test_size=0.2, random_state=0)
```

```python
## Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)

## Random forest
rf = RandomForestRegressor()
rf.fit(X_train, y_train)

la = Lasso()
la.fit(X_train, y_train)

## score Linear Regression
print('Precision of {:.2f} was obtained with Linear Regression model'.format(lr.score(X_test, y_test)))

# score Random Forest Regressor
print('Precision of {:.2f} was obtained with Random Forest Regressor model'.format(rf.score(X_test, y_test)))

## score Lasso
print('Precision of {:.2f} was obtained with Lasso model'.format(la.score(X_test, y_test)))
```

This approach didn't work, so I'm going to add more features to the model to see what happen.

```python
## house_data_mod
house_data_mod = house_data.drop(['id', 'date', 'yr_renovated', 'sqft_living15','sqft_lot15', 'sale_year', 'sqft_basement', 'yr_from_renovation'], axis=1)
#house_data_mod = house_data[['price', 'bathrooms', 'sqft_living', 'view', 'sqft_lot', 'sqft_basement', 'grade', 'zipcode']]
house_data_mod.head()
```

```python
### separating column
target = house_data_mod['price']
feature = house_data_mod.drop(['price'], axis=1)

##splitting the data into 80% for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=0)
```

```python
## setting the scaler
scaler = StandardScaler()

## fitting training set
X_train_std = scaler.fit_transform(X_train)

## apply transform to test set
X_test_std = scaler.transform(X_test)
```

```python
## Linear Regression
lr = LinearRegression()
lr.fit(X_train_std, y_train)
print('Score of {:.2f} was obtained with Linear Regression model'.format(lr.score(X_test_std, y_test)))


## Random forest
rf = RandomForestRegressor()
rf.fit(X_train_std, y_train)
print('Score of {:.2f} was obtained with Random Forest Regressor model'.format(rf.score(X_test_std, y_test)))


## Lasso
la = Lasso()
la.fit(X_train_std, y_train)
print('Score of {:.2f} was obtained with Lasso model'.format(la.score(X_test_std, y_test)))
```

* By adding more features to the model it was possible to improve somewhat the precition capacity getting up to 88% with the *RandomForestRegressor* model.
* Even I did't analize them in the EDA, 'zipcode', 'lat' and 'long' columns help to imporve the prediction.
